```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('ta_restaurants_curated.csv', index_col=0)
data
```

```python
def parse_float(floating_number):
    if floating_number == '':
        return 0.0
    else:
        return float(floating_number)
    
def convert_column(dataset, column_name):
    for i in data[column_name]:
        return parse_float(i)
        
convert_column(data, 'Ranking')
convert_column(data, 'Rating')
convert_column(data, 'Number of Reviews')
```

```python
# How many reviews are there?
# len(data)
print(f"There are {data.size} reviews")
```

```python
# Good vs. Bad Reviews
print(f"There are {(3 < data['Rating']).sum()} bad reviews")
print(f"There are {(3 > data['Rating']).sum()} good reviews")
# 3 < data['Rating']
```

```python
def unique_cities(data):
    unique_set = set()
    for i in data['City']:
        unique_set.add(i)
    return unique_set

u_cities = unique_cities(data)

import operator

cities_dictionary = {}

for i in u_cities:
    cities_dictionary[i] = 0
    
for i in data['City']:
    if i in cities_dictionary.keys():
        cities_dictionary[i] += 1
        
sorted_cities = sorted(cities_dictionary.items(), key=operator.itemgetter(1))

for i in sorted_cities:
    print(i)
```
